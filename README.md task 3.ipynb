# 🏡 Housing Price Prediction using Linear Regression

This project is part of an AI & ML Internship task that focuses on implementing **Simple and Multiple Linear Regression** using Python libraries like **Scikit-learn**, **Pandas**, and **Matplotlib**.

---

## 📌 Objective

To build a Linear Regression model to predict housing prices based on various features using the given dataset.

---

## 📁 Dataset

- Source: [Kaggle - Housing Price Prediction Dataset](https://www.kaggle.com/datasets/harishkumardatalab/housing-price-prediction)
- File used: `Housing.csv`
- Features include: number of bedrooms, bathrooms, area, location, etc.
- Target: `price`

---

## 🛠 Tools & Libraries Used

- Python 🐍
- Pandas 🐼
- NumPy
- Scikit-learn
- Matplotlib & Seaborn

---

## 🔍 Project Workflow

1. **Data Loading**  
   Load the housing dataset using Pandas.

2. **Data Preprocessing**  
   - Handle categorical features using one-hot encoding
   - Check for null values
   - Basic data exploration

3. **Train-Test Split**  
   - 80% training and 20% testing data using `train_test_split`

4. **Model Training**  
   - Fit a Linear Regression model using `sklearn.linear_model.LinearRegression`

5. **Evaluation Metrics**  
   - MAE (Mean Absolute Error)
   - MSE (Mean Squared Error)
   - R² Score

6. **Visualization**  
   - Plot actual vs predicted prices
